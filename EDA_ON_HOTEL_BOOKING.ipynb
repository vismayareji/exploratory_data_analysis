FIND OUTLIERS USING BOXPLOT

A BAR GRAPH SHOWING THE TOP 10 COUNTRIES WITH MOST CANCELED RESERVATIONS OF BOOKING

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\Data science\DATA SET\hotel_booking.csv")
canceled_data=df[df['is_canceled']==1]
country_cancellation=canceled_data['country'].value_counts()
top_10_countries=country_cancellation.head(10)
plt.figure(figsize=(10,6))
top_10_countries.plot(kind='bar',color='orange')
plt.title("top 10 countries with canceled reservation",fontsize=16)
plt.xlabel("country",fontsize=12)
plt.ylabel("number of canceled reservations",fontsize=12)
for i,val in enumerate(top_10_countries):
    plt.text(i,val + 5,str(val),ha='center',fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#14 check the average daily rate of reservation through the graph
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:\Data science\DATA SET\hotel_booking.csv")
df = df.dropna(subset=['adr'])
daily_adr=df.groupby('reservation_status_date')['adr'].mean()

plt.figure(figsize=(12,6))
plt.plot(daily_adr.index,daily_adr.values,color='blue',label='average daily rate(ADR)')
plt.title('average daily rate over time',fontsize=16)
plt.xlabel('date',fontsize=14)
plt.ylabel('ADR(average daily rate)',fontsize=14)
plt.grid(linestyle='--',alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```
